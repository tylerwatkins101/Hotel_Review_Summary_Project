```python
data = pd.read_csv('cleaned_reviews.csv')
```

```python
def full_hotel_review(hotel_name):
    print(hotel_name)
    hotel = data[data.name == hotel_name]
    hotel_review_summarizer.rating_reviews(hotel)
    plt.figure
    hotel_review_summarizer.generate_wordcloud(hotel)
    plt.figure
    hotel_review_summarizer.get_sentiment(hotel)
    hotel_review_summarizer.get_topics(hotel)
```
